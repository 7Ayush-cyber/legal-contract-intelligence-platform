```python
from openai import OpenAI
client = OpenAI()
```

```python
def ask_legal_rag(question):

    retrieved_docs = retriever.invoke(question)

    context = "\n\n".join(
        doc.page_content
        for doc in retrieved_docs
    )

    prompt = f"""
You are a legal contract assistant.

Answer ONLY from the provided context.

If answer cannot be found, say:

Information not found in retrieved contracts.

QUESTION:
{question}

CONTEXT:
{context}
"""

    response = client.responses.create(
        model="openai.gpt-oss-120b",
        input=prompt
    )

    answer = response.output_text

    sources = [
        doc.metadata["source"]
        for doc in retrieved_docs
    ]

    return {
        "question": question,
        "answer": answer,
        "sources": sources
    }
```

```python
evaluation_questions = [

    # Governing Law
    "What is the governing law?",

    # Termination
    "What are the termination rights?",

    # Confidentiality
    "What confidentiality obligations exist?",

    # Indemnification
    "What indemnification obligations exist?",

    # Liability
    "What limitations of liability exist?",

    # Assignment
    "Can either party assign the agreement?",

    # Renewal
    "Under what conditions can the agreement be renewed?",

    # Notices
    "What notice requirements exist?",

    # Payment
    "What payment obligations exist?",

    # Hallucination Tests
    "Who is the CEO of the company?",
    "What is the stock price?",
    "Who won the FIFA World Cup?"
]
```

```python
results = []

for question in evaluation_questions:

    print("="*80)
    print(question)

    output = ask_legal_rag(question)

    results.append(output)

    print(output["answer"])
```

```python
# Create Evaluation DataFrame
df = pd.DataFrame(results)
df.head(10)
```
